```python
import pandas as pd
```

#EX1

```python
weather = pd.read_csv('nyc_weather_2018.csv')
stations = pd.read_csv('weather_stations.csv')
```

```python
weather.head()
```

```python
stations.rename(columns={'id': 'station'}, inplace=True)
stations.head()
```

```python
weather.station.describe()
```

```python
stations.station.describe()
```

```python
weather_snow = weather[weather.datatype=='SNOW']
weather_snow.head()
```

```python
 station_us = stations[stations.name.str.contains('NJ US')]
 station_us.head()
```

```python
 station_info = station_us.merge(weather_snow, on='station')
 station_info.shape
```

```python
station_info.drop_duplicates('station', inplace=True)
station_info.shape
```

```python
station_info.elevation.mean()
```

```python
fb.head()
```

```python
fb.assign(
    volume_pct_change=fb.volume.pct_change(), pct_change_rank=lambda x: \
    x.volume_pct_change.abs().rank(ascending=False)
    ).nsmallest(5, 'pct_change_rank')
```

#EX3

```python
earthquakes = pd.read_csv('earthquakes.csv')
earthquakes.head()
```

```python
earthquakes_ml = earthquakes[earthquakes.magType=='ml']
earthquakes_ml.head()
```

```python
earthquakes_ml = earthquakes_ml.assign(q=pd.cut(earthquakes_ml.mag, [0, 1, 2, 3, 4, 5], labels=['q1', 'q2', 'q3', 'q4', 'q5']))
earthquakes_ml[['mag', 'q']].head()
```

```python
earthquakes_ml.q.value_counts()
```

#EX4

```python
dt = pd.read_csv('nyc_weather_2018.csv', index_col='date', parse_dates=True)
dt.head()
```

```python
dt_snow = dt[dt.datatype=='SNOW']
dt_snow.head()
```

```python
pd.crosstab(
    index=dt_snow.station,
    columns=dt_snow.index.month,
    colnames=['month'],
    values=dt_snow.value,
    aggfunc=lambda x: x.mean(),
    )
```

```python
 pd.crosstab(
    index=dt_snow.station,
    columns=dt_snow.index.month,
    colnames=['month'],
    values=dt_snow.value,
    aggfunc=lambda x: (x > 0).sum(),
    )
```
