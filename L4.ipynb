```python
import pandas as pd
```

#EX1

```python
aapl = pd.read_csv('aapl.csv')
amzn = pd.read_csv('amzn.csv')
fb = pd.read_csv('fb.csv')
goog = pd.read_csv('goog.csv')
nflx = pd.read_csv('nflx.csv')
```

```python
aapl['ticker'] = 'AAPl'
amzn['ticker'] = 'AMZN'
fb['ticker'] = 'FB'
goog['ticker'] = 'GOOG'
nflx['ticker'] = 'NFLX'
```

```python
faang = pd.concat([aapl, amzn, fb, goog, nflx], axis=0)
faang
```

```python
faang.to_csv('faang.csv')
```

```python
faang['volume'] = faang['volume'].astype(int)
```

```python
faang['date'] = pd.to_datetime(faang['date'], errors = 'coerce')
```

```python
faang.sort_values(['date', 'ticker'], axis=0)
faang
```

```python
sort_volume = faang.sort_values('volume', axis=0)
sort_volume
```

#EX2

```python
covid19 = pd.read_csv('covid19_cases.csv')
covid19.head()
```

```python
covid19['date'] = pd.to_datetime(covid19['dateRep'], errors = 'coerce')
covid19.head()
```

```python
covid19.set_index('date',inplace=True)
covid19.head()
```

```python
covid19['countriesAndTerritories'].replace({'United_States_of_America':'USA', 'United_Kingdom':'UK'}, inplace=True)
covid19.head()
```

```python
covid_clean = covid19.loc[covid19['countriesAndTerritories'].isin(["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK",  "USA"])]
covid_clean.head()
```

```python
pivot_covid = covid_clean.pivot(columns="countriesAndTerritories",values="cases").fillna(0)
pivot_covid.head()
```
